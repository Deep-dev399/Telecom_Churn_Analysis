Data Wrangling

```python
print("No. of customers Churning :",df[df['churn']==True].churn.count())  # Total no. of customers churning

#Variable assigning to churn data
churn_df = df[df['churn']==True]

#Toatl no. of unique Pincode
print("No. of unique Pincode :",churn_df.pincode.nunique())
```

```python
# %s of total customer churning
print("Total no. of customers :",df.churn.count() )
perc_churn = (df[df['churn']==True].churn.count()/df.churn.count())*100
print(f"Percentage of customers churning : {round(perc_churn,2)}%")

```

```python
#Statewise customer churning

state_cust_churn = churn_df.groupby(['state'])['churn'].count().reset_index(name='Churn_customer')
print(state_cust_churn.sum())
state_cust_churn
```

```python
# Pincode wise churn count
pin_code_churn_count= churn_df.groupby(['pincode'])['churn'].count().reset_index(name='Counts')
pin_code_churn_count
```

```python
# Poor connectivity by intersection of states by pincode
Pin_state_churn = churn_df.groupby(['pincode'])['state'].unique().reset_index(name='unique states')
Pin_state_churn
```

Visualization of dependent variables

```python
# Chart-1 Dependent Column Value Counts
print(df.churn.value_counts())
print(" ")

# Dependent variables column visualization
df['churn'].value_counts().plot(kind='pie',figsize=(10,6),autopct='%1.1f%%',
startangle=50,
shadow=True,
labels=['Not Churn(%)','Churn(%)'],
colors=['red','green'],
explode=[0.12,0]),
plt.title('Total Percentage of Churn')

# Display Chart
plt.show()
```

# Chart-2

```python
for col in df.describe().columns:
  fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(15,5))
  sns.histplot(df[col],ax=axes[0])
  sns.boxplot(df[col],ax=axes[1],orient='h',color='pink')
  fig.suptitle("Distribution plot of "+ col,fontsize = 15)
  plt.show()
```

```python
# Visualization top 10 churned states
plt.figure(figsize=(18,5))
barl = ((df.groupby(['state'])['churn'].mean()*100).sort_values(ascending=True).reset_index(name='Average True Churn').head(10))
plots = sns.barplot(data = barl, x='state', y='Average True Churn')
for bar in plots.patches:
  plots.annotate(format(bar.get_height(), '.2f'),
                 (bar.get_x() + bar.get_width() / 2,
                  bar.get_height()), ha='center', va='center',
                 size=10, xytext=(0, 5),
                 textcoords='offset points')

plt.title('State with least churn percentage',fontsize=20)
plt.xlabel('State',fontsize=15)
plt.ylabel('Percentage(%)',fontsize=15)
plt.ylim(0,30)
plt.show()
```
